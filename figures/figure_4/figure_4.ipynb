# Figure 4: Novel gene predictions and top selected module features

Figure items:

* 'top_20_novel_genes_scored.csv': List of the top 20 genes (aka false positives) as scored by the final model that are not in the set of known genes (curated, befree, svensson). 
* 'top_64_features.csv': Top 64 modules based on feature scores by ANOVA F-value from sklearn's SelectKBest Feature selector. The dependent variable of ANOVA was whether the gene was in the curated list.

```python
import pandas as pd
```

## Novel gene predictions

```python
all_scores = pd.read_csv('../../results/final_model_svc/all_gene_scores.csv')
```

```python
# 'known' == 1 if the gene is in any of the curated list, the befree list, or svensson list
novel_genes_scores = all_scores[all_scores['known'] == 0]
```

```python
novel_genes_scores.iloc[0:21, :].to_csv('top_20_novel_genes_scored.csv')
novel_genes_scores.iloc[0:21, :]
```

## Top model features

```python
feature_scores = pd.read_csv('../../results/final_model_svc/feature_scores.csv', index_col=0).T
```

```python
top_64 = feature_scores.sort_values('score', ascending=False).iloc[0:64,:]
```

```python
top_64.to_csv('top_64_features.csv')
```
