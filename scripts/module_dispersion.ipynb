# Look at similarities within modules to investigate dispersion of embeddings within modules

```python
from modules import *
import matplotlib.pyplot as plt
import os
import pickle as pkl
from scipy.stats import pearsonr, spearmanr
M = Modules()
```

```python
table = M.gene_module_normalized_scores()
```
